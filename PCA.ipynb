```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import make_blobs

# Step 1: Generate a synthetic dataset
n_samples = 200
random_state = 20
X, y = make_blobs(n_samples=n_samples, n_features=3, random_state=random_state)

# Step 2: Apply PCA
# We want to reduce the dataset to 2 dimensions for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Step 3: Display results
# Original Data (in 3D)
fig = plt.figure(figsize=(16, 8))
ax = fig.add_subplot(121, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y)
ax.set_title("Original Data (3D)")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")

# Data after PCA (in 2D)
ax2 = fig.add_subplot(122)
scatter = ax2.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
ax2.set_title("Data after PCA (2D)")
ax2.set_xlabel("Principal Component 1")
ax2.set_ylabel("Principal Component 2")

# Adding a legend for clusters
legend1 = ax2.legend(*scatter.legend_elements(),
                     loc="lower left", title="Classes")
ax2.add_artist(legend1)

plt.tight_layout()
plt.show()
```
